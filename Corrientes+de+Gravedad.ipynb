```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from pylab import *

#FunciÃ³n para dos ejes verticales
def two_scales(ax1,time,data1,data2,c1,c2):
    ax2=ax1.twinx()
    ax1.plot(time,data1,color=c1)
    ax1.set_xlabel('Tiempo[s]')
    ax1.set_ylabel('Temperatura [Â°C]')
    
    ax2.plot(time,data2,color=c2)
    ax2.set_ylabel('Salinidad (kg/mÂ³)')
    return ax1, ax2


# Serie de tiempo
t=np.arange(1,120,1)

#Temperatura
T=np.loadtxt('temp_sensor1_exp2.txt')

#Salinidad
s=np.loadtxt('fgl_2_calibrate2.txt')


#Generando la grÃ¡fica
figure, ax=plt.subplots()
ax1,ax2=two_scales(ax,t,T,s,'r','b')
ax.set_title('Serie de tiempo')
ax.grid(True)
figure.set_dpi(160) #TamaÃ±o de la imagen



def color_y_axis(ax, color):
    for t in ax.get_yticklabels():
        t.set_color(color)
    return None
color_y_axis(ax1, 'r')
color_y_axis(ax2, 'b')

plt.legend(bbox_to_anchor=(1.1, 1), loc=3, borderaxespad=0.)

plt.show()
```
