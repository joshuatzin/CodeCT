```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
lamerced=pd.DataFrame(pd.read_csv("LaMerced.txt",sep='\t'))
lamerced = lamerced.replace(-99, np.NaN)
#No se toman años con más de 100 NaN en el registro
```

```python
df=pd.DataFrame(lamerced)
plt.figure(figsize=(20,5))
fig = df.boxplot()
fig.set_ylabel("PM2.5 [µg/m³]")
fig.set_title("La Merced ")
df.describe()
```
