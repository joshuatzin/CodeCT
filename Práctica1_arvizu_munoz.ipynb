##### 1. Graficar f(x)=sin(x+1), 2<x<4

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(2,4,0.1)
y=np.sin(x+1)

plt.plot(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('f(x)=sin(x+1)')
plt.show()
```

##### 2. Graficar f(x)=sin(2x) y su derivada en -pi<x<pi en la misma gráfica,incluir toda la información

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-np.pi,np.pi,0.1)
y=np.sin(2*x)
z=2*np.cos(2*x)


plt.plot(x,y,label='sin(2x)')
plt.plot(x,z,label='2cos(2x)')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('f(x)=sin(2x) y su derivada')
plt.legend(loc=0)
plt.show()
```

##### 3.Realizar una gráfica con 4 subgráficas en donde se grafique f(x)=ax^2+bx+c, usando distintos valores de a,b y c

```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint, uniform,random

a=randint(0,1000)
b=randint(0,1000)
c=randint(0,1000)
x=np.arange(1,50,1)

y=a*(x**2)+b*x+c

plt.subplot(141)
plt.plot(x,y)

plt.subplot(142)
plt.plot(x,y)

plt.subplot(143)
plt.plot(x,y)

plt.subplot(144)
plt.plot(x,y)

plt.subplots_adjust(left=0.001,wspace=0.8,top=0.8)

plt.show()
```

##### 4.Graficar la funcion f(x,y)=100(y-x^2)^2 + (1-x)^2 con (x,y) en [-1,1] X [-2,2]

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.arange(-1, 1, 0.25)
Y = np.arange(-2, 2, 0.25)
X, Y = np.meshgrid(X, Y)
Z=100*(Y-X**2)**2+(1-X)**2
surf = ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,linewidth=0, antialiased=False)
plt.show()
```
